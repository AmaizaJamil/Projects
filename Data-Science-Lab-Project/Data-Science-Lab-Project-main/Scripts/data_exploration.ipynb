# Data Exploration: Renewable Energy Readiness and Usage

Importing the libraries and defying the data frame

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("../Datasets_and_maps/merged.csv")

# Clean and standardize column names
df = df.drop(columns=["Unnamed: 0"])
df.columns = df.columns.str.strip().str.replace(" ", "_")
df.head()
```

## Distribution of Key Indicators

The following histograms, enhanced with KDE (Kernel Density Estimation) curves, show how these variables are distributed across all countries in the dataset. This helps identify common value ranges, skewness, and potential outliers.

```python
sns.set_theme(style="whitegrid")
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

sns.histplot(df["Renewable_Energy_score"], bins=15, kde=True, ax=axes[0], color="skyblue")
axes[0].set_title("Distribution of Renewable Energy Score")
axes[0].set_xlabel("Renewable Energy Score")

sns.histplot(df["Renewable_Share"], bins=15, kde=True, ax=axes[1], color="seagreen")
axes[1].set_title("Distribution of Renewable Share")
axes[1].set_xlabel("Renewable Share")

plt.tight_layout()
plt.show()
```

## Grouped Analysis by Income Group

The boxplots below provide a visual summary of the distribution, median, and variability of the Renewable Energy Score indicator and Renewable Share indicator within each group (Low, Lower-middle, Upper-middle, and High income).

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

sns.boxplot(data=df, x="Income_group", y="Renewable_Energy_score", ax=axes[0], palette="Blues", hue="Income_group", legend=False)
axes[0].set_title("Renewable Energy Score by Income Group")
axes[0].set_xlabel("Income Group")
axes[0].set_ylabel("Renewable Energy Score")

sns.boxplot(data=df, x="Income_group", y="Renewable_Share", ax=axes[1], palette="Greens", hue="Income_group", legend=False)
axes[1].set_title("Renewable Share by Income Group")
axes[1].set_xlabel("Income Group")
axes[1].set_ylabel("Renewable Share")

plt.tight_layout()
plt.show()
```
