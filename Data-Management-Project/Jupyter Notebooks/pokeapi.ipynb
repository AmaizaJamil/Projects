# PokeAPI

### Library Import

```python
import requests
import csv
```

### Connection to PokeAPI
Name, Type and Stats retrival


```python
def get_pokemon_data():
    url = "https://pokeapi.co/api/v2/pokemon/"
    pokemon_data = []

    while url:
        response = requests.get(url)
        if response.status_code != 200:
            print(f"Errore nella richiesta: {response.status_code}")
            break

        data = response.json()
        for pokemon in data["results"]:
            pokemon_details = requests.get(pokemon["url"]).json()

            # name
            name = pokemon_details["name"]

            # Tipi
            types = pokemon_details["types"]
            type1 = types[0]["type"]["name"] if len(types) > 0 else None
            type2 = types[1]["type"]["name"] if len(types) > 1 else None

            # Statistiche
            stats = {stat["stat"]["name"]: stat["base_stat"] for stat in pokemon_details["stats"]}
            hp = stats.get("hp", 0)
            attack = stats.get("attack", 0)
            defense = stats.get("defense", 0)
            special_attack = stats.get("special-attack", 0)
            special_defense = stats.get("special-defense", 0)
            speed = stats.get("speed", 0)

            # Aggiungi al dataset
            pokemon_data.append({
                "name": name,
                "type1": type1,
                "type2": type2,
                "hp": hp,
                "attack": attack,
                "defense": defense,
                "special_attack": special_attack,
                "special_defense": special_defense,
                "speed": speed
            })

        # Imposta l'URL successivo
        url = data["next"]
    return pokemon_data
```

### Saving the created dataset in a CSV file

```python
def save_to_csv(pokemon_data, filename="pokemon_data.csv"):
    with open(filename, mode="w", newline="", encoding="utf-8") as file:
        writer = csv.DictWriter(file, fieldnames=["name", "type1", "type2", "hp", "attack", "defense", "special_attack", "special_defense", "speed"])
        writer.writeheader()
        writer.writerows(pokemon_data)
```

### Main

```python
if __name__ == "__main__":
    print("Sto raccogliendo i dati dei Pokémon...")
    data = get_pokemon_data()
    save_to_csv(data)
    print("Dati salvati in 'pokemon_data.csv'.")
```
