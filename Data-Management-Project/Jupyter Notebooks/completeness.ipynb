# COMPLETENESS OF THE DATASETS USED#

### Completeness of the dataset Poke_data 

The code in the first cell analyzes the completeness of the 'poke_data.csv' dataset by calculating the total number
and percentage of missing values (NaN) in each column and assessing the overall completeness
of the entire dataset.

### Completeness of the dataset Players

The code in the second cell evaluates the completeness of the 'players.csv' dataset.
It does this by calculating the total number and percentage of missing values (NaN) in each column.
Additionally, it assesses the overall completeness of the dataset by determining the total number of elements and the proportion of missing values across the entire dataset.


```python
import pandas as pd
import numpy as np


pokemon = pd.read_csv('pokemon_data.csv') 
print (pokemon)

pkmn_column_nan = pd.DataFrame(pokemon.isna().sum(), columns=['Total NaN'])
print(pkmn_column_nan)

# Calculation of the percentage of NaN values for each column, considering that the dataset contains 1302 rows.
pkmn_column_nan['% NaN'] = (pkmn_column_nan['Total NaN']/1302)*100
pkmn_column_nan = pkmn_column_nan.round(decimals=2)
print(pkmn_column_nan)

#completeness of the entire dataset
print(pokemon.size)
print(pokemon.isna().sum().sum())
print(pokemon.isna().sum().sum() / pokemon.size * 100)


```

```python
import pandas as pd
import numpy as np


player = pd.read_csv('players.csv') 
print (player)

pkmn_column_nan = pd.DataFrame(player.isna().sum(), columns=['Total NaN'])
print(pkmn_column_nan)

# Calculation of the percentage of NaN values for each column, considering that the dataset contains 4163 rows.
pkmn_column_nan['% NaN'] = (pkmn_column_nan['Total NaN']/4163)*100
pkmn_column_nan = pkmn_column_nan.round(decimals=2)
print(pkmn_column_nan)

# Completeness of the entire dataset
print(player.size)
print(player.isna().sum().sum())
print(player.isna().sum().sum() / player.size * 100)
```
