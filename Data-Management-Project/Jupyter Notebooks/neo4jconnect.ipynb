# neo4j Connection

### Imports

```python
from neo4j import GraphDatabase
```

### Connection to the server, and verify connectivity
Username and password may be changed if using a differete database

```python
# db connection
URI = "neo4j://localhost:7687"
AUTH = ("neo4j", "DataMan2024")

with GraphDatabase.driver(URI, auth=AUTH) as driver:
    driver.verify_connectivity()
```

### Load cypher file with the instruction to build the graph
If using a different file, change the valure of the "path" variable

```python
# load cypher file
path='neo4j_import.cypher' # insert here the file.cypher path
with open(path, "r",  encoding="utf-8") as file:
    cypher_script=file.read()
```

### Execution
Execute all instruction in the file one by one

```python
# execution
try:
    with driver.session() as session:
        # execute all commands separated by ;
        commands = [cmd.strip() for cmd in cypher_script.split(";") if cmd.strip()]
        for command in commands:
            session.run(command)
            print(f"Eseguito comando: {command}")
finally:
    driver.close()
```

### Close connection and session

```python
# close connection and session
driver = GraphDatabase.driver(URI, auth=AUTH)
session = driver.session(database="neo4j")

session.close()
driver.close()
```
