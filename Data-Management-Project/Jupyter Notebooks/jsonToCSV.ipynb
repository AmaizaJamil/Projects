# PokeDATA - JSON to CSV

### Imports

```python
import json
import pandas as pd
```

### Opening of the file.json downloaded from PokeDATA

```python
# Load the JSON file
with open('players.json', 'r') as file:
    data = json.load(file)
```

### Iterate on the JSON file to extract: rank, player name, and for each pokemon, name, teratype and nature

```python
# Process the JSON data to extract relevant information
csv_data = []

for player in data:
    rank = player.get("placing")
    player_name = player.get("name")
    for pokemon in player.get("decklist", []):
        row = {
            "rank": rank,
            "player_name": player_name,
            "pokemon_name": pokemon.get("name"),
            "teratypes": pokemon.get("teratype"),
            "nature": pokemon.get("ability")  # Assuming "ability" corresponds to the nature of the Pokémon
        }
        csv_data.append(row)
```

### Conversion of the structure into a dataframe, and subsequent saving to a CSV file

```python
# Convert to DataFrame
df = pd.DataFrame(csv_data)

# Save the DataFrame to a CSV file
df.to_csv('players_teams'.csv, index=False)
```
